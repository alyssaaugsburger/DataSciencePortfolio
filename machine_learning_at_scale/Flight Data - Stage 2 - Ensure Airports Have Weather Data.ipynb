```
# Read in file that has IATA and ICAO codes for each airport in our dataset
# Source: https://ourairports.com/data/

# File location and type
file_location = "/FileStore/tables/iata_and_icao_codes.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
iata_and_icao_codes = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(iata_and_icao_codes)
```

```
#Add origin_icao_code column
#Join on clean airline data with code table on ORIGIN
spark.sql("""select t1.year, t1.month, t1.day_of_week, t1.fl_date, t1.op_unique_carrier, t1.origin, t2.icao_code as origin_icao_code, t1.origin_state_abr, t1.dest, t1.dest_state_abr, t1.crs_dep_time, t1.crs_dep_hr, t1.crs_dep_time_2hr, t1.crs_dep_time_2hr_hr, t1.crs_dep_time_3hr, t1.crs_dep_time_3hr_hr, t1.crs_elapsed_time, t1.distance, t1.first_dep, t1.delay from(
    (select * from airline_clean_full) t1
    left join
    (select IATA_CODE, ICAO_CODE from iata_and_icao_codes) t2
    on trim(t1.origin) = t2.iata_code)""").registerTempTable('add_iata_codes_p1')
```

```
spark.sql("""select * from add_iata_codes_p1""").display()
```

```
#Add dest_icao_code column
#Join on clean airline data with code table on DEST
spark.sql("""select t1.year, t1.month, t1.day_of_week, t1.fl_date, t1.op_unique_carrier, t1.origin, t1.origin_icao_code, t1.origin_state_abr, t1.dest, t2.icao_code as dest_icao_code, t1.dest_state_abr, t1.crs_dep_time, t1.crs_dep_hr, t1.crs_dep_time_2hr, t1.crs_dep_time_2hr_hr, t1.crs_dep_time_3hr, t1.crs_dep_time_3hr_hr, t1.crs_elapsed_time, t1.distance, t1.first_dep, t1.delay from(
    (select * from add_iata_codes_p1) t1
    left join
    (select IATA_CODE, ICAO_CODE from iata_and_icao_codes) t2
    on trim(t1.dest) = t2.iata_code)""").registerTempTable('updated_airline_clean_full')
```

```
#We see that there are 4 airports that do not have weather data. Let's inspect these.
#ICAO CODES missing in weather data = 57A, TJPS, KOGS, KXWA

#57A - only has 1 flight
#TJPS (IATA code is PSE) - no weather data for TJPS or PSE - 7854 flights
  # In ponce, puerto rico. could not find a name like ponce in weather data
#KOGS (IATA code is OGS) - no weather data - 1270 flights
  # In Oswegatchie, NY. Could not find a similar name in weather data
#KXWA (IATA code is XWA) - no weather data - 412 flights
  # In Williston, ND
  # Has weather readings for KISN call sign. This is the airport that previously served the area, but it is now XWA


#spark.sql("""select count(fl_date) from updated_airline_clean_full where origin = 'OGS' or dest = 'OGS'""").display()
#spark.sql("""select * from (select distinct right(trim(call_sign),3) as distinct_calls from weather) where distinct_calls = 'OGS' """).display()
```

```
#It looks like we can use the weather readings for call_sign KISN for icao_code KXWA, so let's replace all origin and dest icao codes with KXWA to be KISN
spark.sql("""select year, month, day_of_week, fl_date, op_unique_carrier, origin, 
          case when origin_icao_code = 'KXWA' then 'KISN'
            else origin_icao_code
            end as origin_icao_code, 
          origin_state_abr, dest,
          case when dest_icao_code = 'KXWA' then 'KISN'
            else dest_icao_code
            end as dest_icao_code, 
          dest_state_abr, crs_dep_time, crs_dep_hr, crs_dep_time_2hr, crs_dep_time_2hr_hr, crs_dep_time_3hr, crs_dep_time_3hr_hr, crs_elapsed_time, distance, first_dep, delay
          from
          updated_airline_clean_full""").registerTempTable('updated_airline_clean_full_v2')
```
